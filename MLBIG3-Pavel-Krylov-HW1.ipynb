Первый вариант -> используюя pandas

```python
import pandas as pd
def create_token_pd(text_token):
    text_token_pd=pd.DataFrame(pd.Series(text_token).value_counts().rank(ascending=False,method = 'dense'),\
                                columns=['type rank'])
    text_token_pd['type frequency']=pd.Series(text_token).value_counts()
    return text_token_pd  
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# http://stackoverflow.com/questions/14432557/matplotlib-scatter-plot-with-different-text-at-each-data-point
# http://stackoverflow.com/questions/19068862/how-to-overplot-a-line-on-a-scatter-plot-in-python

def zipf_plot(text_token_pd):
        plt.figure(figsize=(12,6)) 
        x=text_token_pd['type rank']
        y=text_token_pd['type frequency']

        p = np.poly1d(np.polyfit(x, y,4))
        plt.plot(x, y, 'b.')
        plt.plot(x, p(x), 'r-')

        plt.xlabel("type rank", size=15)
        plt.ylabel("type frequency", size=15)
        plt.title("Кривая Ципфа")
        n=12
        for i, txt in enumerate(text_token_pd.index[:n]):
            plt.annotate(txt, (text_token_pd['type rank'][i],text_token_pd['type frequency'][i]),size=13)
```

Расчитаем и построим кривые Ципфа и Хипса для трех разных стилей речи другим способом 

```python
from nltk import FreqDist
def get_n_types_n_tokens(data_token):
    n_types = []
    n_tokens = []
    tokens = []
    fd = FreqDist()
    for i in range(len(data_token)):    
        tokens = data_token[:i]
        fd.update(tokens)
        n_types.append(len(fd))
        n_tokens.append(sum(fd.values()))          
    return n_types,n_tokens
```

```python
n_types1,n_tokens1 = get_n_types_n_tokens(text1_token)
n_types2,n_tokens2 = get_n_types_n_tokens(text2_token)
n_types3,n_tokens3 = get_n_types_n_tokens(text3_token)
```

Заметно достаточно большое время работы даже у сжатого через лемматизацию текста. Т.е. не эффективно отдавать в Mystem.analyze по одному слову. Необходимо скормить сразу весь текст.

Воспользуемся подходом когда в Mystem.analyze отдается сразу весь текст.

```python
%%time
gram_info1 = m.analyze(text1_after_lmtz)
gram_info2 = m.analyze(text2_after_lmtz)
gram_info3 = m.analyze(text3_after_lmtz)
```

Не все что выдал Mystem.lemmatize имеет часть речи, необходимо отфильтрофать пробелы и прочие символы которые не разпознаются и для которых Mystem.analyze выдает пустой лист. Смотри примеры ниже :

```python
gram_info1[5400]
```

```python
gram_info1[5400]["analysis"][0]["gr"].split(',')[0].split('=')[0]
```

```python
gram_info1[2401]
```

```python
gram_info1[2400]
```

```python
def get_POS_tag(gram_info):   
    POS_tag=[]
    for i in range(len(gram_info)):    
        if not gram_info[i]["text"].isspace():
            # если по какой то причине прошли символы которые не отфильтровались выше то для для не будет 
            # определена часть речи и будет пустой list
            if len(gram_info[i]["analysis"])==1:       
                POS_tag.append(gram_info[i]["analysis"][0]["gr"].split(',')[0].split('=')[0])

    fd_POS_tag = FreqDist()
    fd_POS_tag.update(POS_tag)            
    return fd_POS_tag            
```

```python
POS_tag1=get_POS_tag(gram_info1)
POS_tag2=get_POS_tag(gram_info2)
POS_tag3=get_POS_tag(gram_info3)
```

Задание 3. Посчитаем коэффициент корреляции Спирмена для полученных частот частей речи

```python
index, values = zip(*list(POS_tag1.items()))
text_tag1 = pd.DataFrame(list(values),list(index),columns=['деловой стиль'])
index, values = zip(*list(POS_tag2.items()))
text_tag2 = pd.DataFrame(list(values),list(index),columns=['разговорный стиль'])
index, values = zip(*list(POS_tag3.items()))
text_tag3 = pd.DataFrame(list(values),list(index),columns=['художественный стиль'])
```
